- Searched amazon website with search words "Sprider man book"

```python
book_titles = [
    'The Amazing Spider-Man (Marvel: Spider-Man) (Little Golden Book)',
    'This is Spider-Man Level 1 Reader (World of Reading)',
    'World of Reading Super Hero Adventures: Thwip! You Are It!: Level Pre-1',
    "Marvel Spider-man I'm Ready to Read Sound Book - PI Kids",
    'The Story of Spider-Man (Level 2) (World of Reading)',
    'Marvel - Spider-man Me Reader Electronic Reader and 8 Sound Book Library - PI Kids',
    'Marvel Spider-Man Look and Find Activity Book - PI Kids',
    'Marvel Action: Spider-Man: A New Beginning (Book One)',
    '5-Minute Spider-Man Stories (5-Minute Stories)',
    'Spider-Man: Life Story'
    'Miles Morales: Shock Waves (Original Spider-Man Graphic Novel)',
    'Spider-Man: Maximum Venom: Volume 1 (Marvel Spider-Man) (Screen Comix)',
    'Ultimate Sticker Book: Marvel Spider-Man: Into the Spider-Verse',
    'Marvel Super Heroes Spider-man, Avengers, Guardians, and More! - Me Reader Electronic Reader with 8 Book Library',
    "Marvel's Super Hero Adventures Spider-Man: A Tangled Web (Magnetic Hardcover)",
    'Miles Morales (Marvel Spider-Man) (Little Golden Book)',
    'World of Reading: This is Miles Morales',
    "Marvel Super Hero Adventures Buggin' Out!: An Early Chapter Book (Super Hero Adventures Chapter Books, 3)",
    'Marvel Spider-Man: Miles Morales to the Rescue!: Meet the amazing web-slinger! (DK Readers Level 1)'
    "Spider-Man: No Way Home: Spider-Man's Very Strange Day!",
    'Spider-Man: Into the Spider-Verse -The Art of the Movie',
    'Spiderman Coloring Book For Kids: Lots Of Beautiful Illustrations And Patterns For You To Freely Color And Have More Fun.',
    'Spider-Man: Far From Home: Spider-Man Swings Through Europe!',
    'Marvel Action: Spider-Man: Spider-Chase (Book Two)',
    'Marvel Action: Black Panther: Stormy Weather (Book One)',
    'Amazing Spider-Man: Story of Spider-Man (Level 2), The: The Story of Spider-Man (World of Reading: Level 2)']
```

```python
print(" Number of titles:", len(book_titles))
```

```python
# adding stop_words="english" will remove the stop words from the titles.
vect = TfidfVectorizer(min_df=1, stop_words="english")                                                                                                                                                                                                   
tfidf = vect.fit_transform(book_titles)                                                                                                                                                                                                                       
pairwise_similarity = tfidf * tfidf.T

simillarity_array = scipy.sparse.csr_matrix.toarray(pairwise_similarity)
pd.DataFrame.sparse.from_spmatrix(pairwise_similarity)
```

```python
pair_df = pd.DataFrame(simillarity_array)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5, forward=True)
sn.heatmap(pair_df)
plt.show()
```

```python
most_similar = []
most_dissimilar = []
for idx, i in enumerate(simillarity_array):

    list_init = []
    list_idx = []
    for j in i:
        list_init.append(j)
    
    list_sort = list_init[:]
    list_sort.sort()
    maxVal = list_sort[-2]
    minVal = list_sort[0]
    maxidx = list_init.index(maxVal)
    minidx = list_init.index(minVal)
    most_similar.append([idx,maxidx,maxVal])
    most_dissimilar.append([idx,minidx,minVal])
```

```python
print("Most similar titles with similarity score:")
print("------------------------------------------")
max_score = 0
first_idx = 0
second_idx = 0
for comb in most_similar:
    if max_score < comb[2]:
        max_score = comb[2]
        first_idx = comb[0]
        second_idx = comb[1]

print(f"  First title (Indexed at {first_idx}): {book_titles[first_idx]}")
print(f"  Second title (Indexed at {second_idx}): {book_titles[second_idx]}")
print("  Similarity Score:", max_score)

print("\nThere are many dissimilar titles with similarity score of ZERO, below are dissimilar titles")
print("  -------------------------------------------------------------------------------------------")

for comb in most_dissimilar:
    print('\n')
    print(f"  First title (Indexed at {comb[0]}): {book_titles[comb[0]]}")
    print(f"  Second title (Indexed at {comb[1]}): {book_titles[comb[1]]}")
    print("  Similarity Score:", comb[2])      
```

From the similarity score, titles 5 (The Story of Spider-Man (Level 2) (World of Reading)) and 24 (Amazing Spider-Man: Story of Spider-Man (Level 2), The: The Story of Spider-Man (World of Reading: Level 2)) are having highest similarity score of 0.9025. Even though both titles are apart, both are from same series might be reason for highest similarity. Most of the titles are dissimilar with similarity score of ZERO. Titles 1 and 2 are most dissimilar with titles 23 and 24. Based on dissimilarity titles, there is no pattern to the similarity.
